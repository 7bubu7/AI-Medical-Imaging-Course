```
import numpy as np
import matplotlib.pyplot as plt
```

Q1: Create a 1D array with 100 elements and random value (mean = 100, std=5) by **numpy**, plot the 1D array by **plt.plot()**



```
q1 = np.random.normal(100, 5, 100)
plt.plot(q1)
plt.show()
```

Q2: Normalize the 1D array from Q1 by  **(x - xmin) / (xmax - xmin)** and plot the 1D array again

```
q2 = (q1 - np.min(q1)) / (np.max(q1) - np.min(q1))
plt.plot(q2)
```

Q3: Reshape the 1D array from Q2 and plot the 2D array by **plt.imshow**

```
plt.imshow(q2.reshape((10, 10)), cmap='gray')
```
